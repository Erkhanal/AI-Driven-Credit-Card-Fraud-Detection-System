#### Model 1 : Logistic Regression on Imbalanced Data:

```python
from sklearn.model_selection import StratifiedShuffleSplit

# To define the number of splits (kfold) and the split configuration:
kfold = 4
sss = StratifiedShuffleSplit(n_splits=kfold, test_size=0.3, random_state=9487)

# To initialize PowerTransformer:
pt = preprocessing.PowerTransformer(copy=False)

# To fit and transform X using PowerTransformer:
PWTR_X = pt.fit_transform(X)

# To iterate over the splits generated by Stratified Shuffle Split:
for train_index, test_index in sss.split(PWTR_X, y):
    print("TRAIN:", train_index, "TEST:", test_index)

    # Splitting the data into training and test sets based on the indices:
    X_train, X_test = PWTR_X[train_index], PWTR_X[test_index]
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]
```

```python
from sklearn.linear_model import LogisticRegression

# To initialize a logistic regression model:
model_lr = LogisticRegression()

# To fit the logistic regression model to the training data:
model_lr.fit(X_train, y_train)
```

```python
# To predict on the test data using the trained logistic regression model:
y_predicted = model_lr.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, roc_auc_score

# To print classification report:
print('Classification report:\n', classification_report(y_test, y_predicted))

# To print confusion matrix:
print('Confusion matrix:\n', confusion_matrix(y_true=y_test, y_pred=y_predicted))

# To print accuracy score:
print("Logistic Regression Accuracy: ", accuracy_score(y_test, y_predicted))

# To print ROC AUC score:
print('ROC AUC:', roc_auc_score(y_test, y_predicted))
```

```python
import matplotlib.pyplot as plt

def plot_roc_curve(fpr, tpr, roc_auc):

    # To set figure size:
    plt.figure(figsize=(12, 8))

    # To plot the ROC curve:
    plt.plot(fpr, tpr, linewidth=2, label='AUC = %0.3f'% roc_auc, color='b')

    # To plot the diagonal line representing random guessing:
    plt.plot([0, 1], [0, 1], linestyle='--', linewidth=2, color='gray')

    # To set limits for x and y axes:
    plt.xlim([-0.01, 1])
    plt.ylim([0, 1.01])

    # To add a legend to the plot in the lower right corner:
    plt.legend(loc='lower right')

    # To set the title of the plot:
    plt.title('Receiver Operating Characteristic (ROC) Curve', fontsize=14)

    # To set labels for y-axis and x-axis:
    plt.ylabel('True Positive Rate (Sensitivity)', fontsize=14)
    plt.xlabel('False Positive Rate (1 - Specificity)', fontsize=14)

    # To display grid on the plot:
    plt.grid(True)

    # To display the plot:
    plt.show()
```

```python
from sklearn.metrics import roc_curve, roc_auc_score

# To compute ROC curve and ROC area for predictions:
fpr, tpr, thresholds = roc_curve(y_test, y_predicted)
roc_auc = roc_auc_score(y_test, y_predicted)
```

```python
# To plot ROC curve using the computed values:
plt.rcParams['figure.figsize'] = [6, 6]
plot_roc_curve(fpr, tpr, roc_auc)
```
